```python
import pandas as pd
```

```python
movies_df = pd.read_csv("IMDb movies.csv")
movies_df.head()
```

```python
movies_df.drop(columns=['budget', 'usa_gross_income', 'worlwide_gross_income', 'metascore'], inplace=True, axis=1)
movies_df.head()
```

```python
movies_df["reviews_from_users"] = movies_df["reviews_from_users"].fillna(0)
movies_df["reviews_from_critics"] = movies_df["reviews_from_critics"].fillna(0)

movies_df["country"] = movies_df["country"].fillna("UNKNOWN")
movies_df["language"] = movies_df["language"].fillna("UNKNOWN")
movies_df["director"] = movies_df["director"].fillna("UNKNOWN")
movies_df["writer"] = movies_df["writer"].fillna("UNKNOWN")
movies_df["production_company"] = movies_df["production_company"].fillna("UNKNOWN")
movies_df["actors"] = movies_df["actors"].fillna("UNKNOWN")
movies_df["description"] = movies_df["description"].fillna("UNKNOWN")
```
